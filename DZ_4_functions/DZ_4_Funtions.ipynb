```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

#задали декомпозицию кода в виде функции.
def main(docs=documents,dirs=directories):
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            find_name_by_number(docs)
        elif user_input == 's':
            find_dir_by_number(dirs)
        elif user_input == 'l':
            get_fullinfo(docs, dirs)
        elif user_input == 'ads':
            dir_new(dirs)
        elif user_input == 'ds':
            dir_del(dirs)
        elif user_input == 'q':
            print('До свидания!')
            break

#описание функции "Поиск владельца по номеру документа"
def find_name_by_number(docs):
    user_input = input('Введите номер документа (точно с учетом пробелов и других разделителей): ')
    for item in docs:
        if user_input == item['number']:
            return print('Владелец документа: ', item['name'])
    return print('Документ не найден в базе! Повторите ввод!')            
            
#описание функции "Поиск директории (полки) по номеру документа"
def find_dir_by_number(dirs):
    user_input = input('Введите номер документа (точно с учетом пробелов и других разделителей): ')
    for key, item in dirs.items():
        if user_input in item:
            return print(f'Номер полки: {key}')
    return print('Документ не найден в базе! Повторите ввод!')

#описание функции "Вывод полной информации по документам"
def get_fullinfo(docs, dirs):
    for item in docs:
        for dir_key, dir_item in dirs.items():
            if item['number'] in dir_item:
                print(f"№: {item['number']}, тип: {item['type']}, владелец: {item['name']}, номер полки: {dir_key}")
    return

#описание функции "Добавление новой директории (полки)"
def dir_new(dirs):
    user_input = input('Введите номер новой полки: ')
    if user_input in dirs.keys():
        print (f'Такая полка уже существует. Текущий перечень полок:')
        print(*list(dirs.keys()),sep =", ")
    else:
        dirs[user_input] = []
        print('Номер полки добавлен. Текущий перечень полок: ')
        print (*list(dirs.keys()),sep =", ")
    return

#описание функции "Удаление пустой директории (полки)"
def dir_del(dirs):
    user_input = input('Введите номер полки для удаления: ')
    if user_input not in dirs.keys():
        print('Такой полки не существует. Текущий перечень полок: ')
        print(*list(dirs.keys()),sep =", ")
    else:
        for key, val in list(dirs.items()):
            if val != [] and key == user_input:
                print ('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:')
                print(*list(dirs.keys()),sep =", ")
            elif val == [] and key == user_input:
                del dirs[user_input]
                print('Полка удалена. Текущий перечень полок:')
                print(*list(dirs.keys()),sep =", ")
    return

#вызов фукнции
main()
```
