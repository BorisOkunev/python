## 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
# Найти сотрудников с высоким окладом и посчитать статистику их времени на работе.
df_high_salary = df[df['salary'] == 'high']
df_high_salary['average_montly_hours'].describe()
```

Вывод: Сотрудники с высоким окладом работают в среднем 199 часов в месяц.

```python
# Найти сотрудников с низким окладом и посчитать статистику их времени на работе.
df_low_salary = df[df['salary'] == 'low']
df_low_salary['average_montly_hours'].describe()
```

Вывод: Сотрудники с высоким окладом работают в среднем 200 часов в месяц.

Выдвинем гипотезу H0 об отсутствии различий во времени работы сотрудников с высоким и низким доходами. Соответственно, альтернативная гипотеза H1 - существуют различия во времени работы сотрудников с высоким и низким доходами. Проведем статистическое исследование (t-test). Принимаем значимость - 5% (0,05).

```python
import numpy as np
```

```python
X_high = df_high_salary['average_montly_hours'].to_numpy()
```

```python
X_low = df_low_salary['average_montly_hours'].to_numpy()
```

```python
from scipy import stats
```

```python
stats.ttest_ind(X_high,X_low)
```

Вывод: т.к. pvalue = 0,46 - больше уровня значимости 0,05, то гипотеза H0 не отвергается. Таким образом, утверждение, что работники с низким доходом работают больше работников с высоким доходом статистически не подтверждается.

```python
import matplotlib.pyplot as plt
plt.hist(X_low,bins=25)
plt.hist(X_high,bins=25)
```

Справочно: гистограммы распределения времени работы у разных категорий работников.

## 9. Разделить данные на тестовую и обучающую выборки. Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary). Оценить качество модели на тестовой выборки.

```python
df.columns
```

```python
factors = df[['satisfaction_level', 'last_evaluation', 'number_project',
       'average_montly_hours', 'time_spend_company', 'Work_accident',
       'promotion_last_5years']]
factors
```

```python
target = np.array(df['left'])
target
```

```python
# разделяем выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(factors, target, test_size=0.20)
```

```python
# импортируем LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# обучаем данные
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# смотрим разницу факта и прогноза
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
result
```

```python
# расчет точности - отношение верный прогнозов к общему количеству позиций
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(X_test))
```

Ответ: качество модели - 0,77 (77% верных прогнозов)
